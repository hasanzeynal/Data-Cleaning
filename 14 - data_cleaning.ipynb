# The Second Data

```python
second_df = df[['Date.1','Product ID','Customer ID','State ID','Units']]
```

```python
second_df
```

```python
#changing product names
second_df['Product ID'].replace({'Pr_1':'Video Camera',
                  'Pr_2':'Scanner',
                  'Pr_3':'Tablet'},inplace=True)

#changing customers names
second_df['Customer ID'].replace({'Cust_01':'eBay',
                  'Cust_02':'Microsoft',
                  'Cust_03':'IBM',
                  'Cust_04':'Google',
                  'Cust_05':'Twitter',
                  'Cust_06':'Xerox',
                  'Cust_07':'Amazon',
                  'Cust_08':'Intel'},inplace=True)

#changing state id to the name of countries
second_df['State ID'].replace({1:'Azerbaijan',
                  2:'Turkey',
                  3:'Kazakhstan',
                  4:'Iran',
                  5:'Poland',
                  6:'Ukraine',
                  7:'Germany',
                  8:'France',
                  9:'Italy',
                  10:'Spain'},inplace=True)
```

```python
second_df.to_csv('second_df.csv')
```

```python
second_df.head()
```

```python
#creating new column named Sales to calculate daily sales for all products
for j,i in enumerate(second_df['Product ID']):
    if i == 'Tablet':
        second_df['Sales'][j] = second_df['Units'][j] * 28
    if i == 'Video Camera':
        second_df['Sales'][j] = second_df['Units'][j] * 32
    if i == 'Scanner':
        second_df['Sales'][j] = second_df['Units'][j] * 22
```

```python
#creating new column named revenue to calculate daily revenue for all products
for j,i in enumerate(second_df['Product ID']):
    if i == 'Tablet':
        second_df['Revenue'][j] = second_df['Units'][j] * 19
    if i == 'Video Camera':
        second_df['Revenue'][j] = second_df['Units'][j] * 15
    if i == 'Scanner':
        second_df['Revenue'][j] = second_df['Units'][j] * 15
```

```python
second_df.to_csv('second_df.csv')
```
