## some col names is written in "" or there is different only one thin in the first letter of col names like Fitness and fitness and they are seem like different category


```python
for cats in content.category.unique():
    content.replace({cats:cats.lower()},inplace=True)
```

```python
content.category.unique()
```

```python
content['category'] = content['category'].str.replace(r'\W','',regex=True)
```
