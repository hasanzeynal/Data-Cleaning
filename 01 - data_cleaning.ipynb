```python
import pandas as pd
import re
```

```python
dirty_data = pd.read_excel('MO14-Round-1-Dealing-With-Data-Workbook.xlsx',sheet_name='Usage',na_values='',header=None)
dirty_data.head()
```

```python
dirty_data.rename(columns={0:'col_name'},inplace=True)
```

```python
dirty_data['col_name'] = dirty_data['col_name'].str.replace('_',' ')
```

```python
dirty_data
```

```python
dirty_data['meridian'] = dirty_data['col_name'].str.extract('(\d+)')
```

```python
dirty_data['meridian_type'] = dirty_data['col_name'].str.extract('(PM|AM)')
```

```python
dirty_data['col_name'] = dirty_data['col_name'].str.replace('th|st|rd|nd', '')
```

```python
kwh_list = []
for rows in dirty_data.col_name:
    kwh = re.findall(r'\d+',rows)[-2] + '.' + re.findall(r'\d+',rows)[-1]
    kwh_list.append(kwh)
```

```python
dirty_data
```

```python
dirty_data['date'] = dirty_data['col_name'].str.extract('(\d+\-[A-Za-z]+\-\d+)')
```

```python
dirty_data = pd.concat([dirty_data,pd.DataFrame(data = kwh_list,columns=['kwh'])],axis=1)
```

```python
dirty_data
```

```python
dirty_data['date'] = pd.to_datetime(dirty_data['date'], format="%d-%b-%Y")
```

```python
dirty_data['weekday'] = dirty_data['date'].dt.day_name()
```

```python
dirty_data
```

```python
dirty_data['month'] = dirty_data['date'].dt.month_name()
```

```python
dirty_data.drop('col_name',axis=1,inplace=True)
```
